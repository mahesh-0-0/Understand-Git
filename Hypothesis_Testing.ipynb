### 1) Assumption : Something which we assume to be

### 2) Data : Data is that to tell assumption is correct or not (Evidance) / To support or negate the assumption

#### Hypothesis testing is a statistical method that is used to make a statistical decision using experimental data. Hypothesis testing is basically an assumption that we make about a population parameter. It evaluates two mutually exclusive statements about a population to determine which statement is best supported by the sample data.

####  Hypothesis testing is a procedure, based on sample evidence and probability, used to test claims regarding a characteristic of a population.

 - #### **Defining Hypotheses**
  1) Null Hypothesis (H0) : Default assumption or a basic assumption
 made based on the problem knowledge

  2) Alternative Hypothesis (Ha) : It is opposite of Null Hypothesis, i.e. Having evidance data to conclude that default assumption is false

#### P-value : P-value is defined as the Probability of obtaining results at least as extreme as the observed results, under the assumption that the Null Hypothesis (Innocence) is true. Therefore, if p-value is low -> Reject  H0

When using in applications, we compare the p-value with a **threshold value (Î±)**, and if it is less than that, then we reject the $H_0$.

This threshold is known as **significance level**.

#### **Judge Example**

- A person comes into court charged with a crime. A judge must decide whether the person is innocent (null hypothesis) or guilty (alternative hypothesis). Even though the person is charged with the crime, at the beginning of the trial (and until the judge declares otherwise) the accused is assumed to be innocent. Only if overwhelming evidence of the person's guilt can be shown is the judge expected to declare the person guilty--otherwise the person is considered innocent.

  - #### **Errors**
  Two types of Errors

    1) the person is innocent but the judge finds the person guilty (Type I Error) : the null hypothesis is true but the decision based on the testing process is that the null hypothesis should be rejected (False Positive). We call the probability of making a Type-I Error as **alpha (*Î±*)**

    2) the person is guilty but the judge declares the person to be innocent (Type II Error) : the null hypothesis is false but the testing process concludes that it should be accepted (False Negative). We call the probability of making a Type-II Error as **beta (*Î²*)**

    Type I error is usually considered more serious than a Type II error

```python
import numpy as np
from scipy import stats


# Null hypothesis: The population means of fuel efficiency for Engine X and Engine Y are the same.
# Alternative hypothesis: One engine type is more fuel-efficient than the other.

# Define the function to calculate the test statistic and corresponding p-value
def TwoSampZTest(samp_mean_1, samp_mean_2, samp_std_1, samp_std_2, n1, n2):
  # Calculate the test statistic
  denominator = np.sqrt((samp_std_1**2 / n1) + (samp_std_2**2 / n2))
  z_score = (samp_mean_1 - samp_mean_2) / denominator
  return z_score


# Given data
sample_mean_X = 120 # Average fuel efficiency for Group X (Engine X)
sample_mean_Y = 110 # Average fuel efficiency for Group Y (Engine Y)
sample_std_X = 15 # Standard deviation for Group X
sample_std_Y = 12 # Standard deviation for Group Y
significance_level = 0.05
sample_size_X = 30 # Sample size for Group X
sample_size_Y = 35 # Sample size for Group Y


# Calculate the z-score using the function
z_score = TwoSampZTest(sample_mean_X, sample_mean_Y, sample_std_X, sample_std_Y, sample_size_X, sample_size_Y)

# Calculate the two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# A two-tailed z-test considers deviations from the null hypothesis in both positive and negative directions.
# Using abs(z_score) ignores the sign of the deviation, focusing solely on its magnitude, ensuring we consider both directions and calculate the accurate two-tailed p-value.

# Compare the p-value to the significance level
if p_value < significance_level:
  conclusion = "Reject the null hypothesis. Engine Y is more fuel-efficient."
else:
  conclusion = "Fail to reject the null hypothesis. No significant difference in fuel efficiency."

print(f'z-score: {z_score:.4f}')
print(f'p-value: {p_value:.4f}')
print('Conclusion:', conclusion)
```

```python
import numpy as np
import scipy.stats as stats

# Define the data
# Number of satisfied customers
satisfied_customers = 65
# Total number of customers in the sample
total_customers = 100

# Specify the target satisfaction level (null hypothesis)
target_satisfaction = 0.70  # 70%

# Define the hypotheses
# Null Hypothesis (H0): The proportion of satisfied customers is equal to the target satisfaction level.
# Alternative Hypothesis (Ha): The proportion of satisfied customers is not equal to the target satisfaction level.
p = target_satisfaction

# Calculate the test statistic (Z)
# p_hat = sample proportion
p_hat = satisfied_customers / total_customers
n = total_customers
Z = (p_hat - p) / np.sqrt((p * (1 - p)) / n)

# Interpret the test statistic
# Z follows a standard normal distribution. We will calculate the two-tailed p-value next.

#Calculate the p-value
p_value = stats.norm.cdf(np.abs(Z))

# Print the results
print(f"Z-statistic = {Z}")
print(f"P-value = {p_value}")


alpha = 0.05  # Significance level
if p_value < alpha:
    decision = "Reject the null hypothesis"
else:
    decision = "Fail to reject the null hypothesis"

if decision == "Reject the null hypothesis":
    conclusion = "There is significant evidence to suggest that the archers' actual success rate is greater than 70%."
else:
    conclusion = "There is no significant evidence to suggest that the archers' actual success rate is greater than 70%."

# Print the results
print(f"Decision: {decision}")
print(f"Conclusion: {conclusion}")
```

```python
import statsmodels.api as sm
import numpy as np

# Define the data
conversions = np.array([41, 351])  # Number of conversions for Web Page A and Web Page B
visits = np.array([195, 605])  # Number of visits for Web Page A and Web Page B

# Perform the Z-proportions test
z_stat, p_value = sm.stats.proportions_ztest(conversions, visits, alternative='two-sided')

# Print the results
print(f"Z-statistic = {z_stat}")
print(f"P-value = {p_value}")

# Make a decision and conclusion
if p_value < alpha:
    decision = "Reject the null hypothesis"
else:
    decision = "Fail to reject the null hypothesis"

if decision == "Reject the null hypothesis":
    conclusion = "the proportions of Drug A and Drug B are significantly different."
else:
    conclusion = "no significant difference in the proportions of Drug A and Drug B"

# Print the results
print(f"Decision: {decision}")
print(f"Conclusion: {conclusion}")
```

```python
import numpy as np
from scipy.stats import chi2_contingency

# Define the observed frequencies (cross-tabulation) of eye color and hair color
observed = np.array([[20, 94, 84, 17],   # Brown hair
                     [68, 7, 119, 26],     # Black hair
                     [5, 16, 29, 14],     # Blonde hair
                     [15, 10, 54, 14]])    # Red hair

# Perform the Chi-square test
chi2_stat, p_val, dof, expected = chi2_contingency(observed)

# Define the significance level
alpha = 0.01

# Print the test statistic, p-value, and expected frequencies
print("Chi-square statistic:", chi2_stat)
print("P-value:", p_val)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)

# Compare the p-value with the significance level
if p_val < alpha:
    print("Reject the null hypothesis: There is a significant association between eye color and hair color.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between eye color and hair color.")
```
